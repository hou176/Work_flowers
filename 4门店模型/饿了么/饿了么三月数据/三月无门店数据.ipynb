```python
import pandas as pd 
import numpy  as np 
```

```python
data10 = pd.read_excel('一号10月.xlsx')
data8 = pd.read_excel('一号8月.xlsx')
data9 = pd.read_excel('一号9月.xlsx')
```

```python
data10_ = data10[data10['有效订单'] <= 0]
data8_ = data8[data8['有效订单'] <= 0]
data9_ = data8[data8['有效订单'] <= 0]
```

```python
df = pd.concat([data10_,data9_],axis=0)
```

```python
df = pd.concat([df,data8_],axis=0)
```

```python
df = df.drop_duplicates( subset= '门店ID',inplace=False)
```

```python
data = pd.merge(data10_,data9_,on='门店ID')
data = pd.merge(data,data8_,on='门店ID')
```
