```python
import pandas as pd 
import numpy  as np 
pd.set_option('display.max_columns',None)
import  matplotlib.pyplot  as  plt
from pylab import mpl
```

```python
df = pd.read_excel('门店详情.xlsx')
```

```python
df_cj = df[df['商户分层'] == '超级']
df_wm = df[df['商户分层'] == '完美']
df_jk = df[df['商户分层'] == '健康']
df_rm = df[df['商户分层'] == '入门']
df_norm = df[df['商户分层'] == '未达入门']
```

```python
data_9_xx = pd.read_excel('xxdata/九月门店总数据.xlsx')
data_9_xh = pd.read_excel('xhdata/九月门店总数据.xlsx')
data_10_xx = pd.read_excel('xxdata/十月门店总数据.xlsx')
data_10_xh = pd.read_excel('xhdata/十月门店总数据.xlsx')
```

```python
data__9 = data_9_xh[['有效订单']]
data__9.rename(columns = {'有效订单':'有效订单1'},inplace = True)
data_9 = pd.concat([data_9_xx,data__9],axis=1)
```

```python
data_9
```

```python
data__10 = data_10_xh[['有效订单']]
data__10.rename(columns = {'有效订单':'有效订单1'},inplace = True)
data_10 = pd.concat([data_10_xx,data__10],axis=1)
```

```python
data_10
```

```python

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False

x = data_9['日期']
y1 = data_9['有效订单']
y2 = data_9['有效订单1']
plt.figure(figsize=(20,8),dpi=80)

plt.plot(x,y1,color = 'b',label = '其他')
plt.plot(x,y2,color = 'r',label='邂逅')

plt.grid(True, linestyle="--", alpha=1)
plt.xlabel("时间")
plt.ylabel('订单数')
plt.title("9月订单比较", fontsize=20)
plt.legend()
plt.show()
```
