```python
import pandas as pd 
import numpy as  np
pd.set_option('display.max_columns',None)
```

```python
df1 = pd.read_excel('data/壹号花坊.xlsx')
df2 = pd.read_excel('data/壹号花坊详情 .xlsx')
```

```python
data = pd.merge(df1,df2,on='门店ID')
```

```python
data
```

```python
data_T = data[data['总成交额(元)'] > 0]
data_F = data[data['总成交额(元)'] <= 0]
print('总门店数量',data['门店ID'].count())
print('动销门店数量',data_T['门店ID'].count())
print('门店未动销门店',data_F['门店ID'].count())
```

```python
print('有效订单',data_T['有效订单'].sum())
print('流失订单1',data_T['流失订单'].sum())
print('流失订单2',data_F['流失订单'].sum())
```
