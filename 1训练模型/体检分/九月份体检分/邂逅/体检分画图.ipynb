```python
import pandas as pd 
import numpy  as np
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
```

```python
def photo(df):
    df.plot.scatter(x='有效订单数',y='体检分')
    df.plot.scatter(x='入店人数',y='体检分')
    df.plot.scatter(x='平均商家评分',y='体检分')
    df.plot.scatter(x='平均配送评分',y='体检分')
```

```python
list_csv = ['0902','0908','0914','0924','0930']
dataframe = pd.DataFrame()
for i in range(len(list_csv)):
    url = list_csv[i] + '.csv'
    df = pd.read_csv(url,encoding='utf8')
    dataframe =pd.concat([dataframe,df],axis=0)
    
    
```

```python
dataframe.corr()['体检分']
```

```python
dataframe = dataframe.drop(columns=['Unnamed: 0','平均配送评分.1'])
```

```python
photo(dataframe)
```

```python
dataframe_hege = dataframe[dataframe['体检分'] > 60 ]
```

```python
photo(dataframe_hege)
```

```python
x = dataframe[(dataframe['平均商家评分'] > 0)  & (dataframe['体检分'] <65)]
```

```python
x.drop_duplicates(subset='门店ID',inplace=True)
```

```python
x
```

```python
y = pd.read_csv('0930邂逅.csv',encoding = 'gbk')
```

```python
x = pd.merge(x,y,on='门店ID')
```
