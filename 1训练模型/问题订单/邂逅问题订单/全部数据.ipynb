```python
# dataframe导入
def improt_data(lujing,code):
    data = pd.read_excel(lujing,encoding =code,sheet_name= i )
    return data
```

```python

# dataframe传参  此为参数字段名称
# 有多少不同的字段类型病统计输出
def drop_du(data,a):
    df = data.drop_duplicates(subset=a,inplace=False,keep = 'last')
    list_ = list(df[a])
    list_sum = []
    list_sum_ = []
    list__ = []
    for i in range(len(list_)):
        b = list_[i]
        sum_ = data[data[a] == b ]['门店ID'].count()
        dict_ = {}
    
        if sum_ > 1 :
            list__.append(b)
            dict_ = {b:sum_}
            list_sum.append(dict_)
        else:
            list__.append(b)
            dict_ = {b:sum_}
            list_sum_.append(dict_)
        i = i + 1  
    k = [list_sum,list_sum_,list__]
    return k
# 筛选数据dataframe类型传入，一级分类  二级分类   二级分类传参值域
def shaixuan(data,a,b,c):
    data = data[data[a] == b]
    df = data.drop_duplicates(subset=c,inplace=False,keep = 'last')
    list_ = list(df[c])
    list_sum = []
    list_sum_ = []
    list__ = []
    for i in range(len(list_)):
        j = list_[i]
        
        sum_ = data[data[c] == j ]['门店ID'].count()
        dict_ = {}
        if sum_ > 1 :
            list__.append(j)
            dict_ = {j:sum_}
            list_sum.append(dict_)
        else:
            list__.append(j)
            dict_ = {j:sum_}
            list_sum_.append(dict_)
        i = i + 1 
    k = [list_sum,list__,list_sum_]
    return k 

# 简单判断筛选输出新的dataframe
def syn(data,a,b):
    dataframe = pd.DataFrame()
    for i in range(len(b)):
        x = b[i]
        df = data[data[a] == x ]

        dataframe = dataframe.append(df)
    return dataframe
        
```
