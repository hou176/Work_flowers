```python
def create_data(df,data,datatime_0,datatime_1,data_list):
    test = df[['门店','门店ID','城市(城市负责人)']]
    df = df[['门店ID','体检分']]
    df.columns =  ['门店ID',datatime_1]
    data = data[data_list]
    test = pd.merge(test,data,how='left',on='门店ID')
    test = pd.merge(test,df,how='left',on='门店ID')
    test = test.fillna(-1)
    test['value'] = test[datatime_0]-test[datatime_1]
    df = test
    return  df
```

```python
df = create_data(df,data,'12.19','12.20',datalist)
```

```python
df['test20'] = None
df.test20[df.value > 0] = -1
df.test20[df.value == 0] = 0
df.test20[df.value < 0] = 1
df = df.drop(columns=['value'])
```
