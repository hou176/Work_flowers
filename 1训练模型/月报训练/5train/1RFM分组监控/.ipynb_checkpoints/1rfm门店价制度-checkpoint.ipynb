```python
import pandas as pd 
import numpy as np 
pd.set_option('display.max_columns',None)
```

```python
def improt_data(lujing,code):
    data = pd.read_csv(lujing,encoding =code)
    return data
```

```python
lujing = r'C:/Users/Administrator/Datas/爱尚花语/艾尚花语指标模型/test/clean_data.csv'
code = 'utf8'
data = improt_data(lujing,code)
```

```python
data.head()
```

```python
rfm = data.groupby(['商家ID'],as_index=False).agg({'（日均）在线营业时长（h）': 'min',  '复购人数': 'count', '实付交易额': 'sum'})
```

```python
rfm.columns = ['会员ID','r','f','m']
rfm
```

```python
desc_pd = rfm.iloc[:,1:].describe().T
desc_pd
```

```python
r_bins = [-1,11.75,14.75,24]
f_bins = [0,1,1.1,1.2]
m_bins = [-1,0,250,2596]
```

```python
from sklearn.ensemble import RandomForestClassifier # RF库
clf = RandomForestClassifier()
```

```python
clf = clf.fit(rfm[['r','f','m']],rfm['会员ID'])
weights = clf.feature_importances_
print('feature importance:',weights)
```

```python
# RFM分箱得分
rfm['r_score'] = pd.cut(rfm['r'], r_bins, labels=[i for i in range(len(r_bins)-1,0,-1)])  # 计算R得分
rfm['f_score'] = pd.cut(rfm['f'], f_bins, labels=[i+1 for i in range(len(f_bins)-1)])  # 计算F得分
rfm['m_score'] = pd.cut(rfm['m'], m_bins, labels=[i+1 for i in range(len(m_bins)-1)])  # 计算M得分
```

```python
rfm = rfm.apply(np.int32)
```
