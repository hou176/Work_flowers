```python
from sklearn.feature_extraction.text import CountVectorizer
import jieba # 导入关键字提取库
import pandas as pd  # 导入pandas
import newspaper
```

```python
data_name = data.loc[:,'商品分类']
```

```python
list_name = list(data_name)
```

```python
list_name
```

```python
stopwords = ['\t']
# ' ',',','【','】','送','自','购','花','瓶','爆','款','新','款','特','价','+','『','』'
```

```python
clean_line = []
for line in list_name:
    haha = []
    for word in line:
        if word in stopwords:
            continue
        haha.append(str(word))
    haha = ''.join(haha)
    clean_line.append(haha)
    
clean_line
```

```python
text_list = []
for line in clean_line:
    hou = ",".join(jieba.cut(line))
    print(hou)
    text_list.append(hou)
text_list
```

```python
content = ''.join(text_list)
```

```python
content
```

```python
df_content = pd.DataFrame({'content':text_list})
df_content.head()
```
