```python
# 查看曝光的异常值
import matplotlib.pyplot as plt
import matplotlib
from pylab import mpl

data.boxplot(['曝光人数'],showmeans = True)
```

```python
Q2 = np.percentile(data['曝光人数'],25)
print(Q2)
Q3 = np.percentile(data['曝光人数'],75)
print(Q3)
IQR = Q3 - Q2
print(IQR)
Q4 = Q3 + 1.5*IQR
print(Q4)
Q1 = Q3 - 1.5*IQR
print(Q1)
```

```python
data = data[data['曝光人数'] <Q4]
data.boxplot(['曝光人数'],showmeans = True)
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
# data.corr()['曝光人数']
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

x = data[['入店人数','曝光人数','下单人数']]

         
         
wcss = []
for i in range(1, 11): #循环使用不同k测试结果
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_) #inertia簇内误差平方和
print(wcss)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.preprocessing import scale
from sklearn import cluster
model=cluster.KMeans(n_clusters=3,random_state=10)
model.fit(x)
data['cluster']=model.labels_
# data.head(10)
```

```python
data.groupby(['cluster'])['下单人数'].describe()
```

```python
from sklearn import metrics#调用sklearn的metrics库
x_cluster=model.fit_predict(x)#个体与群的距离
score=metrics.silhouette_score(x,x_cluster)#评分越高，个体与群越近；评分越低，个体与群越远
print(score)
```

```python
centers=pd.DataFrame(model.cluster_centers_)
centers
```

```python
# 验证数据的正态分布
from scipy import stats
```

```python
zhengtai = pd.DataFrame(data,columns=['曝光人数'])
```

```python
u = zhengtai['曝光人数'].mean()
std = zhengtai['曝光人数'].std()  
ks_result = stats.kstest(zhengtai['曝光人数'], 'norm', (u, std))
print('ks_res:',ks_result)
p_value = ks_result[1]
```

```python
zhengtai['曝光人数'].plot(kind='kde', grid=True, style='-k', title='密度曲线')
plt.title('destiny')
plt.show()



plt.scatter(zhengtai.index, zhengtai['曝光人数'], color='r', marker='.', alpha=0.5)
plt.grid()
plt.title('分布')
plt.show()
```
