```python
q3 = np.percentile(data_min['入店人数'],75)
q1 = np.percentile(data_min['入店人数'],25)
iqr = q3 - q1
q4 = q3 + 1.5 * iqr
print(q3)
print(q1)
print(q4)
print(iqr)
```

```python
data_min_rd_max = data_min[data_min['入店人数'] > q4]

data_min_rd_max = data_min_rd_max[(data_min_rd_max['曝光人数'] >100 ) & (data_min_rd_max['入店转化率'] > 5)]
data_min_rd_max.sort_values('入店人数',ascending= False)
data_min_rd_max['bilv'] = data_min_rd_max['实付交易额']/data_min_rd_max['原价交易额']
# data_min_rd_max
```
