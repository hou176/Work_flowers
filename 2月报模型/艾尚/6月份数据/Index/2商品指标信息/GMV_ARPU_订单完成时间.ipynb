```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def improt_data(lujing,code):
    data = pd.read_csv(lujing,encoding =code)
    return data
def plt(a):    
   
    q1 = a.quantile(0.25)
    q3 = a.quantile(0.75)
    iqr = q3-q1
    min_ = a.quantile(0)
    q0 = q1 - 1.5*iqr

    q2 = a.quantile(0.5)

    q4 = q3 + 1.5*iqr
    max_ = a.quantile(1)



    print('--------------')
    print('min',min_)
    print('q0:',q0)
    print('q1 :',q1)
    print('q2 :',q2)
    print('q3 :',q3)
    print('q4:',q4)
    print('max :',max_)
    print('--------------')
    list_ = []
    list_.append(min_)
    list_.append(q0)
    list_.append(q1)  
    list_.append(q2)
    list_.append(q3)
    list_.append(q4)
    list_.append(max_)
```
