```python
import pandas as pd 
import numpy as np 
pd.set_option('display.max_columns',None)
```

```python
def improt_data(lujing,code):
    data = pd.read_csv(lujing,encoding =code)
    return data
```

```python
lujing = r'C:/Users/Administrator/Datas/爱尚花语/艾尚花语指标模型/2data/订单数据.csv'
code = 'gbk'
data = improt_data(lujing,code)
```

```python
data
```

```python
GMV = data['商品实付销售额'].sum()
```

```python
count = data['商品实付销售额'].count()
```

```python
ARPU = GMV/count
```

```python
data.info()
```

```python
data['下单时间'] = pd.to_datetime(data['下单时间'],format='%Y-%m-%d')
data['订单完成时间'] = pd.to_datetime(data['订单完成时间'],format='%Y-%m-%d')
data['订单使用时间'] = data['订单完成时间'] - data['下单时间']
```

```python
data['订单使用时间1'] = (data['订单完成时间'] - data['下单时间']).dt.seconds/60
```

```python
colname=['a']
test=pd.DataFrame(columns=colname)


for index, row  in data.iterrows():
#     row['接单时长'] = row['接单时长'].[:-1]
    row  = list(row['接单时长'])
    rows = row[-4:-1]
    rows = ''.join(rows)
    rows = int(rows)
    df_i = pd.DataFrame([rows], columns=colname) # 这个[]很重要
    test = test.append(df_i)
```

```python
data['接单时间（秒）'] = test['a'].values
```
