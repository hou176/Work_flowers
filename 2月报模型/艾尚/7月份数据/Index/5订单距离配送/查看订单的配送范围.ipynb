```python
data.dropna(inplace=True,subset=['建议送达'])
```

```python
data
```

```python
data['下单时间'] = data['日期'].map(str) + data['下单时间'].map(str)
data['建议送达'] = data['日期'].map(str) + data['建议送达'].map(str)
```

```python
data['下单时间']=pd.to_datetime(data['下单时间'],format='%Y/%m/%d/%H:%M:%S')
data['建议送达']=pd.to_datetime(data['建议送达'],format='%Y/%m/%d/%H:%M:%S')
```

```python
data['系统时间'] = data['建议送达'] - data['下单时间']
```

```python
data['系统时间'] = data.系统时间.dt.seconds
```

```python
data['系统时间'] = data['系统时间']  / 60
```

```python
data_min = data[data['订单距离'] <2.4]
data_z = data[(data['订单距离'] > 2.4) & (data['订单距离'] < 4.02)]
data_max =  data[data['订单距离'] >4.02]
```
