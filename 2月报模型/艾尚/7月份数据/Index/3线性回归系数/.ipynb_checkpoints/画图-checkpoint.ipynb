```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def improt_data(lujing,code):
    data = pd.read_excel(lujing,encoding =code)
    return data
lujing = r'C:\Users\Administrator\Datas\艾尚花语\艾尚月季度数据\季度累计数据\1曝光进店.xlsx'
code = 'utf8'
data = improt_data(lujing,code)
```

```python
# 设置图框的大小
fig = plt.figure(figsize=(20,12))
df = data.drop_duplicates(subset='类型',keep = 'last',inplace=False)

```

```python
y7 = data[data['时间'] == 202007]

plt.plot(df['类型'], # x轴数据
         y7['系数（M）'], # y轴数据
         linestyle = '-', # 折线类型
         linewidth = 4, # 折线宽度
         color = 'steelblue', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 3, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='steelblue', # 点的填充色
   ) # 添加标签
y7 = data[data['时间'] == 202007]

plt.plot(df['类型'], # x轴数据
         y7['系数（D）'], # y轴数据
         linestyle = '-', # 折线类型
         linewidth = 2, # 折线宽度
         color = 'steelblue', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 3, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='steelblue', # 点的填充色
     ) # 添加标签
y6 = data[data['时间'] == 202006]

plt.plot(df['类型'], # x轴数据
         y6['系数（M）'], # y轴数据
         linestyle = '-', # 折线类型
         linewidth = 4, # 折线宽度
         color = 'r', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 3, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='steelblue', # 点的填充色
       ) # 添加标签
y6 = data[data['时间'] == 202006]

plt.plot(df['类型'], # x轴数据
         y6['系数（D）'], # y轴数据
         linestyle = '-', # 折线类型
         linewidth = 2, # 折线宽度
         color = 'r', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 3, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='steelblue', # 点的填充色
       ) # 添加标签
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
plt.title('分类门店不同线性斜率')
plt.xlabel('分类')
plt.ylabel('斜率')
plt.legend()
# 显示图形
plt.show()
```
