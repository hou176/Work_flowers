```python
def improt_data(lujing,code):
    data = pd.read_csv(lujing,encoding =code)
    return data
```

```python
lujing = r'C:/Users/Administrator/Datas/爱尚花语/艾尚月季度数据/季度累计数据/data_rfm.csv'
code = 'gbk'
data = improt_data(lujing,code)
```
