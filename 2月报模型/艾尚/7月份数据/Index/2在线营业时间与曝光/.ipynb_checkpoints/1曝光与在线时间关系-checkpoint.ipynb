```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
Q2_data['比率'] = Q2_data['曝光人数'] / Q2_data['（日均）在线营业时长（h）']
Q2_data= Q2_data[Q2_data['（日均）在线营业时长（h）'] > 0]
Q2_data['比率'].mean()
```

```python
Q2_data.loc[ (Q2_data['比率']>=480), 'flag' ] = 1
Q2_data.loc[ (Q2_data['比率']<480), 'flag' ] = 0
Q2_data
```

```python
used_features = [ '（日均）在线营业时长（h）', '曝光人数']
x = Q2_data[used_features].values
y = Q2_data['flag'].values
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
y = lr.predict(x_test)
```

```python
used_features = [ '（日均）在线营业时长（h）', '曝光人数']
x = data[used_features].values
lr_data = lr.predict(x)
data['逻辑分类'] = lr_data
```
