```python
from sklearn.feature_extraction.text import CountVectorizer
import jieba 
import newspaper
```

```python
df_name = df.loc[:,'商品分类']
```

```python
df_list = list(df_name)
```

```python
df_list
```

```python
text = []
for line in df_list:
    i = ','.join(jieba.cut(line))
    text.append(i)
text
```

```python
stopwords = pd.read_csv('1.txt' , index_col=False,sep='\t',names = ['h'],encoding='gbk')
```

```python
stopwords.head()
```

```python
def drop_stopwords(contents,stopwords):
    contents_clean = []
    all_words = []
    for line in contents:
        line_clean = []
        for word in line:
            print(word)
            if word == '\t':
                continue
            line_clean.append(word)
            all_words.append(str(word))
        contents_clean.append(line_clean)
    return contents_clean,all_words    
```

```python
df_content = pd.DataFrame({'content':text})
df_content.head()
```

```python
contents = df_content.content.values.tolist()
stopwords = stopwords.h.values.tolist()
contents_clean,all_words = drop_stopwords(contents,stopwords)
```
