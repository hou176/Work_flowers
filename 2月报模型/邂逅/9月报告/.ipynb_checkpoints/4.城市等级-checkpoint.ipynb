```python
df_M = pd.read_csv('九月份门店.csv')
```

```python
df_M = df_M.sort_values(by='有效订单数',ascending = False)
```

```python
df_city = pd.read_csv('城市与等级.csv')
df_city = df_city[['城市','城市等级']]
# df_city.to_csv('城市与等级.csv')
```

```python
df_M_city = pd.merge(df_M,df_city,on='城市')
```

```python
df_M_city
```

```python
city_level_list = ['S', 'A', 'B1', 'B2', 'C1', 'C2', 'D1', 'D2', 'D3','0']
```

```python
city_level_list_two = [['有效订单数','实付交易额','曝光人数','入店人数','下单人数'],['商家ID']]
```

```python
k = pd.DataFrame(city_level_list,columns=['城市等级'])
for j in city_level_list_two[0]:
    list_ = []
    for i in city_level_list:
        city_level_dict = []
        a = df_M_city[df_M_city['城市等级'] == i]
        b = a[j].sum()
        list_.append(b)
    city_level_dataframe = pd.DataFrame(list_,columns=[j])
    k = pd.concat([k,city_level_dataframe],axis=1)
```

```python
for j in city_level_list_two[1]:
    list_ = []
    for i in city_level_list:
        city_level_dict = []
        a = df_M_city[df_M_city['城市等级'] == i]
        b = a[j].count()
        list_.append(b)
    city_level_dataframe = pd.DataFrame(list_,columns=['门店数量'])
    k = pd.concat([k,city_level_dataframe],axis=1)
```

```python
k['入店转化率'] = k['入店人数'] / k['曝光人数'] * 100
k['下单转化率'] = k['下单人数'] / k['入店人数'] * 100
# k.to_csv('城市等级门店.csv')
```
